### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each plot
x = np.linspace(0, 10, 10)
y = x**2  # Line plot data

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]  # Bar chart data

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]  # Scatter plot data

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]  # Pie chart data
colors = ['blue', 'red', 'green', 'orange']

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot (Top-Left)
axs[0, 0].plot(x, y, color='b', linewidth=2)
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y = X")
axs[0, 0].grid(True, linestyle="--", alpha=0.7)

# Bar Chart (Top-Right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter Plot (Bottom-Left)
axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].grid(True, linestyle="--", alpha=0.7)

# Pie Chart (Bottom-Right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True)
axs[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
